```python
x = [10, 20, 5, 10, 30, 40,10, 20, 5, 10, 30, 40,50, 40, 7, 20, 40, 50]
alpha = 3
kappa = 0.5
```

```python
x = np.asarray(x, dtype=float)
```

```python
xa = np.power(np.clip(x, 0, None), alpha)
```

np.clip(x, 0, None) clamps any negative spend values to 0. Spend can't be negative in reality, but if a numerical glitch upstream produced a small negative number, (-0.001)**1.6 is undefined in real numbers. Clipping makes the function robust.

np.power(..., alpha) raises every element to the power alpha. This is xαx^\alpha
xα — the numerator (and part of the denominator) of the Hill formula. Storing it in xa ("x to the alpha") is a small optimization: the formula uses xαx^\alpha
xα twice, so compute it once.
